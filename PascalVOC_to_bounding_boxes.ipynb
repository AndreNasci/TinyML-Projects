# Converts PASCAL to bounding-boxes.labels

```python
from os import listdir
from os.path import isfile, join

# Path dos arquivos a serem listados
path = "/src"

xml_files = [f for f in listdir(path) if isfile(join(path, f))]
lista_xml = [arq for arq in xml_files if arq.lower().endswith(".xml")]
```

```python
import xml.etree.ElementTree as ET

files = {}
for aux in lista_xml:
  tree = ET.parse(join(path, aux))
  root = tree.getroot()


  file_name = root[1].text
  print("Processing", file_name)
  print()

  boxes = []


  for box in [i+6 for i in range(9)]:
    item = {}
    item['label'] = root[box][0].text
    xmin = int(root[box][4][0].text)
    ymin = int(root[box][4][1].text)
    xmax = int(root[box][4][2].text)
    ymax = int(root[box][4][3].text)
    item['x'] = xmin
    item['y'] = ymin
    item['width'] = xmax - xmin
    item['height'] = ymax- ymin
    boxes.append(item)


  files[file_name] = boxes
```
